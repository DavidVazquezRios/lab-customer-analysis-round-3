```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
data.shape
```

```python
data.describe(include = "all")
```

```python
# Show a plot of the total number of responses.
sns.displot(data['Response'])
plt.show()
```

```python
int_response= lambda x: 1 if x=='Yes' else 0

data['Int Response']=list(map(int_response, data['Response']))
data.groupby('Sales Channel')['Int Response'].mean().plot(kind='bar')
plt.ylabel('Rate Response')
plt.show()
```

```python
data.head()
```

```python
#Show a plot of the response rate by the total claim amount
data.groupby(round(data['Total Claim Amount']/100, 0).astype(int))['Int Response'].mean().plot(kind='bar')
plt.xlabel('Claim Amount x 100')
plt.ylabel('Response Rate')
plt.show()
```

```python
data.head()
```

```python
# Show a plot of the response rate by income
data.groupby(round(data['Income']/10000, 0).astype(int))['Int Response'].mean().plot(kind='bar')
plt.xlabel('Income x 10000')
plt.ylabel('Response Rate')
plt.show()
```
